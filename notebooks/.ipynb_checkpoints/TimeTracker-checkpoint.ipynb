```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/Users/giosferrazza/Desktop/sleep-spending-predictor/data/export_16-02-2026-to-17-02-2026.csv')
df.head()
```

```python
# Convert time columns to timedelta
time_cols = ['Sleeping', 'Leisure', 'Projects', 'Interruption', 'Work', 
             'Eating', 'Commute', 'Studying', 'Gaming', 'Hygiene', 'Exercise']

for col in time_cols:
    df[col] = pd.to_timedelta(df[col])

df.dtypes
```

```python
# Convert to hours as float
for col in time_cols:
    df[col] = df[col].dt.total_seconds() / 3600

df
```

```python
# Set date as index
# df['date'] = pd.to_datetime(df['date']) # Converts the date into something Pandas understands
# df.set_index('date', inplace=True) # Makes the date the row label

# Plot
df[time_cols].plot(kind='bar', figsize=(12, 6)) # Plot columns, type bar, fig size
plt.title('Time Spent Per Activity') # Title of chart
plt.ylabel('Hours') # Y label
plt.xlabel('Date') # X label
plt.xticks(ticks=[0, 1], labels=['Feb 16', 'Feb 17'], rotation=0)
plt.legend(loc='upper right') #legend location
plt.tight_layout() # Prevents label cut off
plt.show() # Actually displays the chart
```
